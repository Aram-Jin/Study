```python
# backtest
```

```python
def get_return_df(df, N=1, log=False):
    if log:
        return np.log(df/df.shift(N)).iloc[N-1:].fillna(0)
    else:
        return df.pct_change(N, fill_method=None).iloc[N-1].fillna(0)
```

```python
def get_cum_return_df(return_df, log=False):
    if log:
        return np.exp(return_df.cumsum())
    else:
        return (1+return_df).cumprod()
```

```python
df['ma_short'] = df['close'].rolling(5).mean()
df['ma_long'] = df['close'].rolling(20).mean()
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df = df[['close', 'ma_short', 'ma_long']]
```

```python
df[-300:].plot(figsize=(10,5))
```

```python
# 골든크로스 - 단기이평이 장기이평을 뚫고 올라가는 경우 => 매수
# 데드크로스 -                       깨고 내려가는 경우 => 매도
```

```python
import numpy as np
```

```python
df.loc[:, 'position'] = np.where(df['ma_short'] > df['ma_long'], 1, 0)
```

```python
exit_index = df[(df['position'] - df['position'].shift()) == -1].index
exit_index
```

```python
long_index = df[(df['position'] - df['position'].shift()) == 1].index
long_index
```

```python
ax = df.drop(['position'], axis=1).plot(figsize=(15,5))
df.loc[exit_index, 'ma_short'].plot(ax=ax, marker='o', linestyle='none', color='red')
df.loc[long_index, 'ma_short'].plot(ax=ax, marker='o', linestyle='none', color='green')
```

```python
df.loc[:, 'rtn'] = get_return_df(df['close'], log=True)
df
```

```python
df.loc[:, 'strategy_rtn'] = df['position'].shift()*df['rtn'].fillna(0)
```

```python
df
```

```python
df.loc[:, 'cum_rtn'] = get_cum_return_df(df['rtn'], log=True)
df.loc[:, 'cum_strategy_rtn'] = get_cum_return_df(df['strategy_rtn'], log=True)
```
