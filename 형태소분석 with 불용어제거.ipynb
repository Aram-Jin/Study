```python
import pandas as pd
import numpy as np 
import re    
from konlpy.tag import Twitter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
from konlpy.tag import Okt
import itertools
```

```python
# 'Unnamed: 0 없애면서 읽기

df = pd.read_csv("C:\\Python\\env01\\LuvvD\\LuvData.csv",index_col = 0)
df.head()
```

```python
df["plaintext"] = df.plaintext.fillna(value=np.nan) 
df = df[pd.notnull(df["plaintext"])]
df.head()
```

```python
df.plaintext.isnull().sum()
```

```python

def remove_html(text): # html tag지우기 
    pattern = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')  # 태그 지우기  & nbsb..지우기 
    return pattern.sub(r"",str(text))
    
def remove_digit(text):  # 숫자 지우기 
    pattern =re.compile("[0-9]|")  
    return pattern.sub(r"",str(text))

def remove_specific(text): # text에 있는 특이한? 문자들 지우기
    pattern=re.compile("\.|\/+|\●+|\\n+")     
    return pattern.sub(r"",str(text))  

def backword_func(text):  # 후방탐색으로 특정 문자 검색, 상담, 진단 만 get하깅
    pattern = re.compile("(?<=검색|상담|진단).+")
    strs = "".join(pattern.findall(text))
    return strs[20:]  # 요 부분은 .. 변경이 될 수 있음 
```

```python
df["cleaned"] = df["plaintext"].apply(remove_html) 
df["cleaned"] = df["cleaned"].apply(remove_digit) 
df["cleaned"] = df["cleaned"].apply(remove_specific)
df["cleaned"] = df["cleaned"].apply(backword_func) 
df.head()
```

```python
stop_words = pd.read_csv("C:\\Python\\env01\\LuvvD\\Korean_Stop_words.csv",header=None)
stop_words
```

```python
stop_words = stop_words.astype({0: str})    # type object 로 바꿔주고 
stop_words.rename(columns = {0:'words'}, inplace = True)
stop_words
```

```python
stop_words_list = stop_words.words.tolist()
```

```python
# 불용어 제거하기 

okt = Okt()

df1 = df.copy()
sample = df1[:100]
final = []
for x in sample.cleaned:
    word_tokens = okt.morphs(x)
    new = list(itertools.chain(word_tokens))
    final.append(new)

word_list = [word for lis in final for word in lis]
result = [word for word in word_list if not word in stop_words_list]
result
```

```python
undefined_words = []
analysis=[]
twit = Twitter()
for word in result:
    morphs = twit.pos(word)
    if len(morphs) == 1:
        res = "".join("@".join(tup) for tup in morphs)
        analysis.append(res)
    else:
        undefined_words.append(morphs)
analysis

undefined_words
```
